# Preprocesado de los audios

En este notebook, se hará el primer paso de preprocesamiento de los audios con 3 procesos distintos:

- Eliminar el silencio del principio y final de los audios.
- Eliminar el ruido de fondo
- Normalizar el volumen

### Normalizar el volumen

```python
# Importamos las librerias necesarias
import os
from pydub import AudioSegment


def NormalizarVolumen(carpeta):
    """
    Normalizar el volumen de todos los audios.

    Parámetros:
    - carpeta: directorio donde se encuentran todos los audios a tratar.

    Retorna:
    - Los audios normalizados.
    """
    for archivo in os.listdir(carpeta):
        # Comprobar el formato de audio
        if archivo.endswith(".wav"):

            try:
                # Cargar audio
                audio = AudioSegment.from_file(os.path.join(carpeta, archivo)

                # Normalizar el volumen 
                audio_normalizado = audio.normalize()

            # Excepción con mensaje de salida del archivo que ha no se ha podido normalizar
            except Exception as e:
                print(f"No se pudo cargar el archivo {archivo}: {e}")

            #Exportar el sudio normalizado en la carpeta indicada
            audio_normalizado.export(os.path.join(carpeta, archivo), format="wav")
```

```python
# Carpeta de audios
carpeta = "Audios_preprocesados/"

# Llamamos a la función
NormalizarVolumen(carpeta)
```

### Eliminar principio y final

```python
# Importamos las librerias necesarias
import os
from pydub import AudioSegment
import numpy as np

def EliminarSilencion(carpeta, umbral):
    """
    Eliminar tanto el silencio del inicio como del final de todos los audios.

    Parámetros:
    - carpeta: directorio donde se encuentran todos los audios a tratar.
    - umbral: cifra para determinar los segundos mínimos que deben durar los audios tras el preprocesado.

    Retorna:
    - Los audios preprocesados.
    """
    for i in os.listdir(carpeta):

        try:
            # Cargar el archivo en cada caso
            audio = AudioSegment.from_file(os.path.join(carpeta, i))
            print(i)

            # Obtener la matriz del audio
            matriz_audio = np.array(audio.get_array_of_samples())

            # Índice del primer valor que no sea un silencio
            inicio_sin_s = np.argmax(matriz_audio > 0.00)

            # Índice del último valor que no sea un silencio
            final_sin_s = len(matriz_audio) - np.argmax(matriz_audio[::-1] > 0.00)

            # Calcular la duración del audio original y del audio sin silencios
            duracion = len(audio)
            duracion_sin_s = final_sin_s - inicio_sin_s

            # Comprobar si la diferencia de duración es menor que el umbral marcado
            if abs(duracion - duracion_sin_s) < umbral:
                # Si la diferencia de duración es menor que el umbral, guardar el audio original
                audio.export(os.path.join(carpeta, "original_" + i), format="wav")
            else:
                # Si la diferencia de duración es mayor que el umbral, recortar el audio y guardarlo en la misma carpeta en formato .wav
                audio_sin_s = audio[inicio_sin_s:final_sin_s]
                audio_sin_s.export(os.path.join(carpeta, i), format="wav")

        except Exception as e:
            print(f"No se pudo cargar el archivo {archivo}: {e}")
```

```python
# Carpeta de audios
carpeta = "Audios_preprocesados/"

# Marcamos un umbral para decidir si guardar el audio original o el audio recortado
umbral = 100 # en milisegundos

# Llamamos a la función
EliminarSilencion(carpeta, umbral)
```
